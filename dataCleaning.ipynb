```python
sns.histplot(data['college_gpa'], kde=True)
plt.title("College GPA Distribution")
plt.show()
```

```python
print(data.columns)
```

```python
data['coding_score'] = pd.to_numeric(data['coding_score'], errors='coerce')
data['placed'] = data['placed'].astype(str)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=data, x='placed', y='coding_score')
plt.title("Coding Score vs Placement")
plt.show()
```

```python
sns.boxplot(x='placed', y='internships', data=data)
plt.title("Internships vs Placement")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(data[['college_gpa','coding_score','internships','projects','placed']], hue='placed')
plt.show()
```

```python
sns.barplot(x='internships', y='placed', data=data)
plt.title("Placement Rate by Internships")
plt.show()
```

```python
sns.boxplot(x='placed', y='technical_skills', data=data)
plt.title("Technical Skills vs Placement")
plt.show()
```

```python
features = ['college_gpa','coding_score','technical_skills','communication_skills']

for i in features:
    plt.figure()
    sns.boxplot(x='placed', y=i, data=data)
    plt.title(i + " vs Placement")
    plt.show()
```

```python
data['experience_score'] = data['internships'] + data['projects'] + data['hackathons']
```

```python
sns.boxplot(x='placed', y='experience_score', data=data)
plt.title("Experience Score vs Placement")
plt.show()
```
